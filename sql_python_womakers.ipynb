```python
import sqlite3

#importando bibliotecas 
import pandas as pd
```

```python
#criando conexão com o banco de dados SQLite

conn = sqlite3.connect('Aula5_Womakers')
cursor = conn.cursor()
```

```python
tabela_alunos = """CREATE TABLE alunos(
'id' INTEGER ,
'nome' VARCHAR(255),
'idade' INTEGER,
'curso' VARCHAR(255)
);"""

cursor.execute(tabela_alunos)
```

```python
tabela_cliente = """CREATE TABLE clientes(
'id' integer PRIMARY KEY ,
'nome' VARCHAR(255),
'idade' INTEGER,
'saldo' FLOAT
);"""

#sqlite não tem AUTO_INCREMENT

cursor.execute(tabela_cliente)

conn.commit()
```

```python
inserindo_clientes = [('Maria', 35, 5000.00),
                      ('João', 28, 7500.00),
                      ('Ana', 45, 10000.00),
                      ('Pedro', 40, 12000.00),
                      ('Carla', 33, 8500.00),
                      ('Lucas', 29, 6000.00),
                      ('Mariana', 38, 9500.00),
                       ('Paulo', 42, 11000.00)]
# Comando SQL para INSERT 
comando_sql = "INSERT INTO clientes (nome, idade, saldo) VALUES (?, ?, ?)"

# Executando o comando para inserir os dados
cursor.executemany(comando_sql, inserindo_clientes)

# Commit das alterações no banco de dados
conn.commit()
```

```python
#SELECT IDADE > 30

df = pd.read_sql_query("""SELECT nome, idade FROM clientes
	WHERE idade > 30""", conn)
df
```

```python
#SELECT MEDIA SALDO

df = pd.read_sql_query("""SELECT AVG(saldo) FROM clientes
""", conn)
df
```

```python
#SELECT MAXIMO SALDO

df = pd.read_sql_query("""SELECT MAX(saldo) FROM clientes""",conn)
df
```
